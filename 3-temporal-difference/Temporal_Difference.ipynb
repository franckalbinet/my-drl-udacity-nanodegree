### Part 1: TD Control: Sarsa

In this section, you will write your own implementation of the Sarsa control algorithm.

Your algorithm has four arguments:
- `env`: This is an instance of an OpenAI Gym environment.
- `num_episodes`: This is the number of episodes that are generated through agent-environment interaction.
- `alpha`: This is the step-size parameter for the update step.
- `gamma`: This is the discount rate.  It must be a value between 0 and 1, inclusive (default value: `1`).

The algorithm returns as output:
- `Q`: This is a dictionary (of one-dimensional arrays) where `Q[s][a]` is the estimated action value corresponding to state `s` and action `a`.

Please complete the function in the code cell below.

(_Feel free to define additional functions to help you to organize your code._)

````python
def epsilon_greedy_probs(q_s, epsilon):
    """Get epsilon greedy probs for given state

    Take a Q table entry for a single state and returns chosen/drawn action from pmf

    Args:
        q_s (list): action values for a particular state
        epsilon (float): level of "greed" betwee 0 and 1 

    Returns:
        list: pmf (probability mass function) of possible actions
    """
    nA = len(q_s)
    idx_max = np.argmax(q_s)
    return np.array([(1 - epsilon) + epsilon/nA if a == idx_max else epsilon/nA for a in range(nA)])
````

````python
def epsilon_greedy_action(q_s, epsilon):
    """ Return a random action based on epsilon greedy probs """
    probs = epsilon_greedy_probs(q_s, epsilon)
    return np.random.choice(np.arange(len(probs)), p=probs) 
````

````python
def update_Q(Qsa, Qsa_next, reward, alpha, gamma):
    """ updates the action-value function estimate using the most recent time step """
    return Qsa + (alpha * (reward + (gamma * Qsa_next) - Qsa))
````

````python
def sarsa(env, num_episodes, alpha, gamma=1.0):
    # Initialize action-value function (empty dictionary of arrays)
    Q = defaultdict(lambda: np.zeros(env.nA))
    
    # Initialize epsilon and define decay_rate
    epsilon = 0.005
    decay_rate = 1e-5
    
    # Initialize performance monitor
    # Loop over episodes
    for i_episode in range(1, num_episodes+1):
        # monitor progress
        if i_episode % 100 == 0:
            print("\rEpisode {}/{}".format(i_episode, num_episodes), end="")
            sys.stdout.flush()   

        # Reset the environment
        s = env.reset()

        while True:
            # Get epsilon-greedy action
            a = epsilon_greedy_action(Q[s], epsilon) 

            # Take actions and get next state, reward and status (done or not)
            s_next, r_next, done, _ = env.step(a)
            
            # Draw next action from e-greedy strategy
            a_next = epsilon_greedy_action(Q[s_next], epsilon)
            
            # Update the Q table
            Q[s][a] = update_Q(Q[s][a], Q[s_next][a_next], r_next, alpha, gamma)
                
            s = s_next
        
            if done:
                break
                
        # Exponential decay for epsilon
        epsilon = epsilon * np.exp(-decay_rate * i_episode)
        
    return Q
````

````python
# obtain the estimated optimal policy and corresponding action-value function
Q_sarsa = sarsa(env, 5000, .01)

# print the estimated optimal policy
policy_sarsa = np.array([np.argmax(Q_sarsa[key]) if key in Q_sarsa else -1 for key in np.arange(48)]).reshape(4,12)
check_test.run_check('td_control_check', policy_sarsa)
print("\nEstimated Optimal Policy (UP = 0, RIGHT = 1, DOWN = 2, LEFT = 3, N/A = -1):")
print(policy_sarsa)

# plot the estimated optimal state-value function
V_sarsa = ([np.max(Q_sarsa[key]) if key in Q_sarsa else 0 for key in np.arange(48)])
plot_values(V_sarsa)
````

````python
def q_learning(env, num_episodes, alpha, gamma=1.0):
    # initialize empty dictionary of arrays
    Q = defaultdict(lambda: np.zeros(env.nA))
    
    # Initialize epsilon and define decay_rate
    epsilon = 0.005
    decay_rate = 1e-5
    
    # loop over episodes
    for i_episode in range(1, num_episodes+1):
        # monitor progress
        if i_episode % 100 == 0:
            print("\rEpisode {}/{}".format(i_episode, num_episodes), end="")
            sys.stdout.flush()
        
        # Reset the environment
        s = env.reset()
        
        while True:
            # Get epsilon-greedy action
            a = epsilon_greedy_action(Q[s], epsilon) 

            # Take actions and get next state, reward and status (done or not)
            s_next, r_next, done, _ = env.step(a)
            
            # Draw next action from e-greedy strategy
            a_next_max = np.argmax(Q[s_next])
                
            # Update the Q table
            Q[s][a] = update_Q(Q[s][a], Q[s_next][a_next_max], r_next, alpha, gamma)
                
            s = s_next
        
            if done:
                break
                
        # Exponential decay for epsilon
        epsilon = epsilon * np.exp(-decay_rate * i_episode)
        
    return Q
````

### Part 3: TD Control: Expected Sarsa

In this section, you will write your own implementation of the Expected Sarsa control algorithm.

Your algorithm has four arguments:
- `env`: This is an instance of an OpenAI Gym environment.
- `num_episodes`: This is the number of episodes that are generated through agent-environment interaction.
- `alpha`: This is the step-size parameter for the update step.

The algorithm returns as output:

Please complete the function in the code cell below.

(_Feel free to define additional functions to help you to organize your code._)

````python
def expected_sarsa(env, num_episodes, alpha, gamma=1.0):
    # initialize empty dictionary of arrays
    Q = defaultdict(lambda: np.zeros(env.nA))
    
    # Initialize epsilon and define decay_rate
    epsilon = 0.005
    decay_rate = 1e-5
    
    # loop over episodes
    for i_episode in range(1, num_episodes+1):
        # monitor progress
        if i_episode % 100 == 0:
            print("\rEpisode {}/{}".format(i_episode, num_episodes), end="")
            sys.stdout.flush()
        
        # Reset the environment
        s = env.reset()
        
        while True:
            # Get epsilon-greedy action
            a = epsilon_greedy_action(Q[s], epsilon) 

            # Take actions and get next state, reward and status (done or not)
            s_next, r_next, done, _ = env.step(a)
            
            probs = epsilon_greedy_probs(Q[s_next], epsilon)
            
            # Update the Q table
            Q[s][a] = update_Q(Q[s][a], np.dot(probs, Q[s_next]), r_next, alpha, gamma)
            
            s = s_next
        
            if done:
                break    
        
        # Exponential decay for epsilon
        epsilon = epsilon * np.exp(-decay_rate * i_episode)
        
    return Q
````
